```python
import os
import json
import xml.etree.ElementTree as ET
import PIL
import torch
import torchvision.transforms.functional as F
import numpy as np
import random
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
output_path = r"/content/gdrive/MyDrive/JSONdata1"
Dataset_path = r"/content/gdrive/MyDrive/Normal_Insulators"
annotation_path=r"/content/gdrive/MyDrive/Normal_Insulators/labels/"
list_of_files = sorted( filter( lambda x: os.path.isfile(os.path.join(annotation_path, x)),
                        os.listdir(annotation_path) ) )
for file_name in list_of_files:
    #print(file_name,end=" ")
   def parse_annotation(annotation_path):
       tree = ET.parse(annotation_path)
       root = tree.getroot()
       print(root)
       boxes = list()
       labels = list()
       difficulties = list()
       for object in root.iter("object"):
          difficult = int(object.find("difficult").text == "1")
          label = object.find("name").text.lower().strip()
          bbox =  object.find("bndbox")
          xmin = int(bbox.find("xmin").text)
          ymin = int(bbox.find("ymin").text)
          xmax = int(bbox.find("xmax").text)
          ymax = int(bbox.find("ymax").text)
          boxes.append([xmin, ymin, xmax, ymax])
          labels.append(label)
          difficulties.append(difficult)
       print({"boxes": boxes, "labels": labels, "difficulties": difficulties})
       return{"boxes": boxes, "labels": labels, "difficulties": difficulties}
   parse_annotation(annotation_path+file_name)

   def create_json_data(Dataset_path, output_path):
    '''
        Create json file saved data
       Dataset_path: Path to normal insulators folder
        output_path: Path to save folder
    '''

    train_images = list()
    train_objects = list()

    with open(os.path.join(r"/content/gdrive/MyDrive/Normal_Insulators/labels/0049.xml")) as f:
        ids = f.read().splitlines()
    for id in ids:
        #Get object, image in XML file
        #object_path = os.path.join(voc_path, "Annotations", id + ".xml")
        object_path = os.path.join(Dataset_path,"labels", id + ".xml")
        objects = parse_annotation(object_path)
        if len(objects) == 0:
           continue
        train_objects.append(objects)
        #image_path = voc_path + "/JPEGImages/"+id+".jpg"
        image_path =Dataset_path +"images"+id+".jpg"
        print(train_objects)
        train_images.append(image_path)
    assert len(train_objects) == len(train_images)
    with open(os.path.join(output_path, "TRAIN_images.json"), "w") as j:
        json.dump(train_images, j)

    with open(os.path.join(output_path, "TRAIN_objects.json"), "w") as j:
        json.dump(train_objects, j)
    test_images = list()
    test_objects = list()

    with open(os.path.join(r"/content/gdrive/MyDrive/Normal_Insulators/labels/1714.xml")) as f:
        ids = f.read().splitlines()

    for id in ids:
       object_path = os.path.join(Dataset_path,"labels", id + ".xml")
       objects = parse_annotation(object_path)
       if len(objects) == 0:
            continue
       test_objects.append(objects)
       test_image_path = Dataset_path + "images" +id+".jpg"
       print(test_objects)
       test_images.append(test_image_path)

    assert len(test_images) == len(test_objects)
    with open(os.path.join(output_path, "TEST_images.json"), "w") as j:
         json.dump(test_images, j)
    with open(os.path.join(output_path, "TEST_objects.json"), "w") as j:
         json.dump(test_objects, j)
    create_json_data(Dataset_path,output_path)
```
